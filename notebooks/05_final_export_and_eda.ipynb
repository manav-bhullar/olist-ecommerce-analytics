# Phase 5: Final Export for Tableau

In this final notebook, we prepare the dataset for our Tableau Public dashboard by ensuring all KPIs (RFM Segments, Delivery Delays, etc.) are correctly formatted.

```python
import pandas as pd

df = pd.read_csv('../data/processed/merged_olist_data.csv')
print(f"Final dataset contains {df.shape[0]:,} rows and {df.shape[1]} columns.")
```

## 1. Select Columns for Tableau
We don't need all 40 columns. We will trim down the dataset to only what's required for the dashboard to optimize performance.

```python
tableau_cols = [
    'order_id', 'customer_unique_id', 'seller_id', 'product_id',
    'order_purchase_timestamp', 'price', 'freight_value',
    'customer_state', 'customer_city',
    'seller_state', 'seller_city',
    'product_category_name',
    'review_score',
    'delivery_delay_days', 'is_delayed',
    'Segment' # RFM Segment
]

df_tableau = df[[col for col in tableau_cols if col in df.columns]].copy()

# Ensure date is properly formatted
df_tableau['order_purchase_timestamp'] = pd.to_datetime(df_tableau['order_purchase_timestamp']).dt.date

df_tableau.head()
```

## 2. Export to CSV

```python
df_tableau.to_csv('../data/processed/tableau_export.csv', index=False)
print("tableau_export.csv generated successfully for Tableau Public!")
```
